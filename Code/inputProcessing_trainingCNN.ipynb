**FRAME EXTRACTION CODE**

```python
video_count = 0

input_loc = "/content/drive/MyDrive/LSAvideos/all"
output_loc = "/content/drive/MyDrive/LSAvideos/Train_Frames"

os.chdir(input_loc)

print("Input Loaction: ", input_loc)
print("Output Location: ", output_loc)

def segmentation(image):
  result = image.copy()
  
  lower1 = np.array([0, 120, 0])
  upper1 = np.array([140, 255, 100])
  
  lower2 = np.array([25, 0, 75])
  upper2 = np.array([180, 38, 255])
  
  lower_mask = cv2.inRange(image, lower1, upper1)
  upper_mask = cv2.inRange(image, lower2, upper2)
  
  mask = cv2.bitwise_or(lower_mask, upper_mask)
  result = cv2.bitwise_and(result, result, mask=mask)
  
  return result

videos = os.listdir(input_loc)
print ("Number of videos: ", len(videos))

for video in tqdm(videos, unit='videos', ascii=True):
    
  video_count = video_count + 1
  last_frame = None
  cap = cv2.VideoCapture(video)
  frameCount = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  
  count = 0

  while count < 180:
    ret, frame = cap.read()
    
    if ret == False:
        break
    
    frame = segmentation(frame)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    last_frame = frame

    frame_name = video.split(".")[0] 
    frame_name = frame_name + "_" + str(count + 1) + ".jpg"
    
    cv2.imwrite(os.path.join(output_loc , frame_name), frame)
    count = count + 1

  while count < 180:
    frame_name = video.split(".")[0] 
    frame_name = frame_name + "_" + str(count + 1) + ".jpg"
    
    cv2.imwrite(os.path.join(output_loc , frame_name), last_frame)
    count = count + 1


  cap.release()
  cv2.destroyAllWindows()
```

```python
def split_list(alist, wanted_parts=1):
    length = len(alist)
    return [ alist[i*length // wanted_parts: (i+1)*length // wanted_parts] 
             for i in range(wanted_parts) ]

train_img_name_vector_A, train_img_name_vector_B = split_list(train_img_name_vector, 2)
```

```python
print(len(train_img_name_vector_A))
print(len(train_img_name_vector_B))
```

```python
image_model = tf.keras.applications.InceptionV3(include_top = False, weights = 'imagenet', pooling='avg')

new_input = image_model.input
hidden_layer = image_model.layers[-1].output

image_features_extract_model = tf.keras.Model(new_input, hidden_layer)
image_features_extract_model.summary()
```

```python
def load_image(image_path):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, (299, 299))
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    return img, image_path
```

**PART A - CNN For First 27000 Frames**

```python
cnn_train = sorted(set(train_img_name_vector_A))

image_dataset = tf.data.Dataset.from_tensor_slices(cnn_train)
image_dataset = image_dataset.map(load_image, num_parallel_calls=tf.data.experimental.AUTOTUNE).batch(64)
```

```python
sample_output_loc = "/content/drive/MyDrive/LSAvideos/CNN_Train_Output_Part1/"
os.chdir(sample_output_loc)

for img, path in tqdm(image_dataset):
  batch_features = image_features_extract_model(img)
  
  for bf, p in zip(batch_features, path):
    path_of_feature = p.numpy().decode("utf-8")
    np.save(path_of_feature, bf.numpy())
```
